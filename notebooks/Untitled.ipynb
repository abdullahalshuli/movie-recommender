```python
import pandas as pd

# Load the datasets
movies_df = pd.read_csv('../data/tmdb_5000_movies.csv')
credits_df = pd.read_csv('../data/tmdb_5000_credits.csv')

# Merge on title
movies_df = movies_df.merge(credits_df, on='title')
movies_df.head()
```

```python
movies_df = movies_df[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies_df.dropna(inplace=True)
movies_df.head()
```

```python
import ast

def convert(obj):
    return [i['name'] for i in ast.literal_eval(obj)]

def get_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return i['name']
    return ''

movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df['cast'] = movies_df['cast'].apply(lambda x: convert(x)[:3])  # top 3 actors
movies_df['crew'] = movies_df['crew'].apply(get_director)
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x: x.split())
movies_df['crew'] = movies_df['crew'].apply(lambda x: [x])

movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
new_df = movies_df[['movie_id', 'title', 'tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

vectors.shape
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    movie = movie.lower()
    if movie not in new_df['title'].str.lower().values:
        print("Movie not found in dataset.")
        return
    
    index = new_df[new_df['title'].str.lower() == movie].index[0]
    distances = list(enumerate(similarity[index]))
    movies = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]

    print(f"\nTop 5 movies similar to '{new_df.iloc[index].title}':")
    for i in movies:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
import pickle

# Save processed movie data and similarity matrix
pickle.dump(new_df, open('../model/movies.pkl', 'wb'))
pickle.dump(similarity, open('../model/similarity.pkl', 'wb'))
```
